```python
import numpy as np
import pandas as pd
import pandas_ta as ta
from ta import add_all_ta_features
from ta.utils import dropna
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from bourse_refs_api.models import StockHistory
```

```python
df = pd.DataFrame(StockHistory.objects.all().values())
df = df.sort_values('date').reset_index(drop=True)
df = dropna(df)
```

```python
df
```

```python
df = df.sort_values(['stock_id', 'date'])
```

```python
df = df.rename(columns={
    'max_price': 'high',
    'min_price': 'low',
    'last_deal_price': 'close',
    'first_price': 'open'
})
```

```python
def add_cdl_patterns(group):
    try:
        return pd.concat([group, group.ta.cdl_pattern()], axis=1)
    except:
        return pd.DataFrame()

df.groupby('stock_id').apply(add_cdl_patterns)
```
